```python
import pandas as pd 
import csv
import folium

# Я заранее подготовила файлы с адресами и координатами
veg = pd.read_csv('https://raw.githubusercontent.com/Levittt/eco_map/master/shops.csv')
bat = pd.read_csv('https://raw.githubusercontent.com/Levittt/eco_map/master/batteries.csv')
containers = pd.read_csv('https://raw.githubusercontent.com/Levittt/eco_map/master/containers.csv')
pd.core.frame.DataFrame

from folium import Map, FeatureGroup, Marker, LayerControl

map = folium.Map(location=[53.39458965465114, 58.980445861816406], zoom_start = 9, tiles = "OpenStreetMap")

fg = FeatureGroup(name='Сдать батарейки') 
fg2 = FeatureGroup(name='Эко магазины')
fg3 = FeatureGroup(name='Сдать пластиковые бутылки')  

for lat, lon, address in zip(bat['lat'],bat['lon'], bat['address']):
    #Layer('Сдать батарейки').add_to(map)
    folium.Marker(location=[lat, lon], tooltip=str(address), popup=str(address), icon=folium.Icon(color = 'gray')).add_to(fg)
for lat, lon, address in zip(veg['lat'],veg['lon'], veg['address']):
    #Layer('Эко магазины').add_to(map)
    folium.Marker(location=[lat, lon], tooltip=str(address), popup=str(address), icon=folium.Icon(color = 'green')).add_to(fg2)
for lat, lon, address in zip(containers['lat'],containers['lon'], containers['address']):
    #Layer('Сдать пластиковые бутылки').add_to(map)
    folium.Marker(location=[lat, lon], tooltip=str(address), popup=str(address), icon=folium.Icon(color = 'pink')).add_to(fg3)

fg.add_to(map)
fg2.add_to(map)
fg3.add_to(map)

folium.LayerControl(collapsed=False).add_to(map)
map.save("eco_map.html")
map
```
